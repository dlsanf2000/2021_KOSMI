### 데이터 불러오기

```python
df_raw = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/KOSMI/heart.csv') #Set Dir
```

```python
df = df_raw.copy()
```

### 결측치 확인

```python
df.isnull().sum()
```

### 정규화

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
standardScaler.fit(df[['age']])
df[['age']] = standardScaler.transform(df[['age']])

standardScaler.fit(df[['trestbps']])
df[['trestbps']] = standardScaler.transform(df[['trestbps']])

standardScaler.fit(df[['chol']])
df[['chol']] = standardScaler.transform(df[['chol']])

standardScaler.fit(df[['thalach']])
df[['thalach']] = standardScaler.transform(df[['thalach']])

standardScaler.fit(df[['oldpeak']])
df[['oldpeak']] = standardScaler.transform(df[['oldpeak']])
```

### 원핫인코딩

```python
pd.get_dummies(df['sex'])
pd.get_dummies(df['cp'])
pd.get_dummies(df['fbs'])
pd.get_dummies(df['restecg'])
pd.get_dummies(df['exang'])
pd.get_dummies(df['slope'])
pd.get_dummies(df['ca'])
pd.get_dummies(df['thal'])
```

### 타겟변수 설정

```python
target_col = 'target'
target = df[target_col]
```

### features 변수들 설정

```python
features = df[['age','sex','cp', 'trestbps','chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
```

### test/train 셋 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size = 0.3, random_state=3
)
```

### 의사결정트리

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
DTree = DecisionTreeClassifier(random_state = 0)
DTree.fit(X_train, y_train)
print(DTree.score(X_test, y_test))

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report

y_pred = DTree.predict(X_test)

print(classification_report(y_test,y_pred))

print("Precision : ", precision_score(y_test, y_pred,average='macro'))
print("Recall : ", recall_score(y_test, y_pred,average='macro'))
print("F1-Score : ", f1_score(y_test, y_pred,average='macro'))
print("Accuracy : ", accuracy_score(y_test, y_pred))
```

### 랜덤포레스트

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=10)
RF.fit(X_train, y_train)

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report

y_pred = RF.predict(X_test)

print("Precision : ", precision_score(y_test, y_pred,average='macro'))
print("Recall : ", recall_score(y_test, y_pred,average='macro'))
print("F1-Score : ", f1_score(y_test, y_pred,average='macro'))
print("Accuracy : ", accuracy_score(y_test, y_pred))
```

### GBM(Gradient Boosting Machine)

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier(random_state=0)
gb.fit(X_train, y_train)


y_pred = gb.predict(X_test)
print(classification_report(y_test,y_pred))

print("Precision : ", precision_score(y_test, y_pred,average='macro'))
print("Recall : ", recall_score(y_test, y_pred,average='macro'))
print("F1-Score : ", f1_score(y_test, y_pred,average='macro'))
print("Accuracy : ", accuracy_score(y_test, y_pred))
```

### XGBoost

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test)

print(classification_report(y_test,y_pred))
print("Precision : ", precision_score(y_test, y_pred,average='macro'))
print("Recall : ", recall_score(y_test, y_pred,average='macro'))
print("F1-Score : ", f1_score(y_test, y_pred,average='macro'))
print("Accuracy : ", accuracy_score(y_test, y_pred))
```
